### Missing imputation

#### Generate some data

```python
import secrets
from causomic.causal_model.LVM import LVM
from causomic.simulation.proteomics_simulator import simulate_data
from causomic.data_analysis.proteomics_data_processor import dataProcess
from causomic.simulation.example_graphs import signaling_network

from sklearn.impute import KNNImputer

import pyro
```

```python
seed = 349267299
```

```python
seed = secrets.randbelow(2**31 - 1)
np.random.seed(seed)
print(f"Sampled random seed: {seed}")
print("Loading signaling network...")
sn = signaling_network(add_independent_nodes=True)

print("Simulating data with missing values...")
data = simulate_data(
    sn["Networkx"], 
    coefficients=sn["Coefficients"], 
    mnar_missing_param=[-5.5, 0.3],  # Missing not at random parameters
    add_feature_var=True, 
    n=50, 
    seed=seed
)
data["Feature_data"]["Intensity"] = data["Feature_data"]["Obs_Intensity"]

print("Processing feature-level data to protein-level...")
summarized_data = dataProcess(
    data["Feature_data"], 
    normalization=False, 
    feature_selection="All",
    summarization_method="TMP",
    MBimpute=False,
    sim_data=True
)

# Remove external interventions for validation
summarized_data = summarized_data.loc[:, [
    col for col in summarized_data.columns if col not in ["IGF", "EGF"]
]]
```

```python
summarized_data.isna().sum()
```

```python
# Clear Pyro parameter store for fresh inference
pyro.clear_param_store()

# Fit Bayesian causal model
lvm = LVM(backend="pyro",
    num_steps=10000,
    patience=10,
    initial_lr=.001,
    verbose=True)
lvm.fit(summarized_data, sn["causomic"])
```

```python
imputer = KNNImputer(n_neighbors=3, keep_empty_features=True)
obs_data_eliator = pd.DataFrame(
    imputer.fit_transform(summarized_data), 
    columns=summarized_data.columns
)
```

#### Fit models

```python
x = "Ras"
z = "Raf"

model =lvm
imp_data = model.imputed_data
X_data = imp_data.loc[imp_data["protein"] == x]
Z_data = imp_data.loc[imp_data["protein"] == z]

X_backdoor_color = np.where(
    (X_data['imp_mean'].isna().values & Z_data['imp_mean'].isna().values), 
    "blue", 
    np.where((X_data['intensity'].isna().values \
              & Z_data['intensity'].isna().values), 
             "red", "orange"))

X_data = np.where(
    X_data['imp_mean'].isna(),
    X_data['intensity'], 
    X_data['imp_mean'])

Z_data = np.where(
    Z_data['imp_mean'].isna(),
    Z_data['intensity'], 
    Z_data['imp_mean'])


fig, ax = plt.subplots(1, 3, figsize=(14, 5), constrained_layout=True)

eliator_data = obs_data_eliator

from scipy.stats import linregress

# Define a list of scatter plot configurations
temp = summarized_data.loc[:, [x, z]].dropna()
plots = [
    (temp.loc[:, x], 
     temp.loc[:, z], "blue",
     "Observed data", x, z),
    (X_data, 
    Z_data, 
     X_backdoor_color, "Bayesian imputation", x,z),
    (eliator_data[x], eliator_data[z], 
     X_backdoor_color, "KNN imputation", x, z),
]

for i, (x, y, color, title, xlabel, ylabel) in enumerate(plots):

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    print(title, slope)
    line_x = np.linspace(x.min()-1, x.max()+1, 100)
    line_y = slope * line_x + intercept

    ax[i].scatter(x, y, color=color,
                  edgecolor='k', s=80, alpha=0.8)
    ax[i].plot(line_x, line_y, color='red', linestyle='--', 
               linewidth=2)
 
    ax[i].set_title(title, fontsize=16, fontweight='bold')
    # ax[i].set_xlabel(xlabel, fontsize=14)
    # ax[i].set_ylabel(ylabel, fontsize=14)
    ax[i].tick_params(axis='both', which='major', labelsize=14)
    ax[i].grid(True, linestyle='--', alpha=0.3)

    # ax[i].set_xlim(3, 9.5)
    # ax[i].set_ylim(1, 12)

fig.suptitle("Imputation Comparison", fontsize=18, fontweight='bold')
fig.supxlabel("Ras", fontsize=16, fontweight='bold')
fig.supylabel("Erk", fontsize=16, fontweight='bold')

plt.show()
```

```python
sn["Coefficients"]["Raf"]
```

```python
int1 = {"Ras": 5}  # Low intervention
int2 = {"Ras": 7}  # High intervention  
outcome = "Erk"        # Outcome

lvm.intervention(int1, outcome)
causomic_int_low = lvm.intervention_samples

# Perform intervention under second condition (normalize intervention value)
lvm.intervention(int2, outcome)
causomic_int_high = lvm.intervention_samples

# Calculate ATE
causomic_int_high.mean() - causomic_int_low.mean()
```

```python
from y0.dsl import Variable
from eliater.regression import summary_statistics

eliator_int_low = summary_statistics(
    sn["y0"], obs_data_eliator,
    treatments={Variable(list(int1.keys())[0])},
    outcome=Variable(outcome),
    interventions={
        Variable(list(int1.keys())[0]): list(int1.values())[0]
    }
)

# Estimate effect under second intervention
eliator_int_high = summary_statistics(
    sn["y0"], obs_data_eliator,
    treatments={Variable(list(int2.keys())[0])},
    outcome=Variable(outcome),
    interventions={
        Variable(list(int2.keys())[0]): list(int2.values())[0]
    }
)

# Calculate ATE
eliator_int_high.mean - eliator_int_low.mean
```

```python
# Plot customization
fig, ax = plt.subplots(1, 4, figsize=(18, 6), sharey=True)

colors = {"Frequentist": "#E69F00", "Bayesian": "#56B4E9"}
replicates = [20, 50, 100, 250]

for i, key in enumerate(replicates):
    data = priors[priors["Replicates"] == key].reset_index(drop=True)

    # Plot ground truth
    ax[i].axvline(data.loc[0, "Ground_truth"], color="black", 
                  linestyle="dashed", lw=2, label="Ground Truth")

    # Plot histograms
    ax[i].hist(data["Eliator"], alpha=0.8, bins=15, 
               color=colors["Frequentist"], label="Frequentist")
    ax[i].hist(data["MScausality"], alpha=0.8, bins=10, 
               color=colors["Bayesian"], label="Bayesian")

    # Titles and formatting
    ax[i].set_title(f"{key} Replicates", fontsize=20, fontweight='bold')
    ax[i].grid(axis='x', linestyle='--', alpha=0.6)
    ax[i].tick_params(axis='both', which='major', labelsize=12)

plt.setp(ax, xlim=(-2.5, 4.5), yticks=[])

# Shared x-axis and y-axis labels
fig.supxlabel("Log$_{2}$ Fold Change", fontsize=18, y=0.05)
fig.supylabel("Frequency", fontsize=18, x=0.005)

# Shared legend outside the grid
handles, labels = ax[0].get_legend_handles_labels()
# fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 1.02), ncol=3, fontsize=14, frameon=False)

# Add overall title
fig.suptitle("ATE Simulation Comparison - Informed Priors", 
             fontsize=24, fontweight='bold')

# Adjust layout
fig.tight_layout()
fig.subplots_adjust(top=0.85, bottom=0.15)

plt.show()
```

```python
priors.loc[:, "Eliator_SE"] = (priors.loc[:, "Eliator"] - \
                                   priors.loc[:, "Ground_truth"])**2
priors.loc[:, "MScausality_SE"] = (priors.loc[:, "MScausality"] - \
                                   priors.loc[:, "Ground_truth"])**2
priors.groupby(["Replicates"])[["Eliator_SE", "MScausality_SE"]].mean()
```
